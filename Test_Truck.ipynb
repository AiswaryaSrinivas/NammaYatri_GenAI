```python
## Truck Filling 
import pandas as pd
data = {
    "SKU": ["SKU1", "SKU2", "SKU3","SKU4"],
    "MinQty": [5, 10, 15,4],
    "MaxQty": [12, 15, 17,6],
    "AvailableQty": [10, 30, 40,11],
    "Multiplier": ["4X", "5X", "3X+2X","4X"],
    "Score":[0.5,0.3,0.1,0.2]
}
dat=pd.DataFrame(data)
```

```python
multiplier_data={"4X":[4]*10,"5X":[5]*10,"3X+2X":[3,2]*5}
multiplier_data
```

```python
def split_min_qty_into_equal_rows(min_qty):
    # Ensure we have enough rows for the minimum quantity
    
    # Distribute 1 unit per row
    return [1 for _ in range(min_qty)]
```

```python
rows_df=pd.DataFrame()
dat=dat.sort_values(by="Score",ascending=False).reset_index(drop=True)
dat['order']=dat.index
dat
```

```python
max_skus=max(dat['order'])
max_skus
```

```python
df_list=[]
for i in range(0,max_skus+1):
    temp_df=dat[dat['order']==i]
    min_rows=split_min_qty_into_equal_rows(temp_df['MinQty'].tolist()[0])
    df_1=pd.DataFrame()
    df_1['QTY']=min_rows
    df_1['SKU']=temp_df['SKU'].tolist()[0]
    df_1['AvailableQty']=temp_df['AvailableQty'].tolist()[0]
    df_1['Multiplier']=temp_df['Multiplier'].tolist()[0]
    df_1['Rank']=temp_df['order'].tolist()[0]

    df_1['Qty_Type']="Min_QTY"
    df_list.append(df_1)
df_1=pd.concat(df_list)
df_1
```

```python
df_list_2=[]


for i in range(0,max_skus):
    temp_df=dat[dat['order']==i]
    temp_df['Remaining_Qty']=temp_df['MaxQty'] - temp_df['MinQty']
    min_rows=split_min_qty_into_equal_rows(temp_df['Remaining_Qty'].tolist()[0])
    df=pd.DataFrame()
    df['QTY']=min_rows
    df['SKU']=temp_df['SKU'].tolist()[0]
    df['AvailableQty']=temp_df['AvailableQty'].tolist()[0]
    df['Multiplier']=temp_df['Multiplier'].tolist()[0]
    df['Rank']=temp_df['order'].tolist()[0]

    df['Qty_Type']="Max_QTY"
    df_list_2.append(df)
df_2=pd.concat(df_list_2)
```

```python
df=pd.concat([df_1,df_2])
df['index']=df.reset_index(drop=True).index
df
```

```python
def get_inventory_remaining(df,sku):
    dat=df[df['SKU']==sku].sort_values(by='index',ascending=True)
    dat['prev_inventory_used_truck']=dat.reset_index(drop=True).index
    dat['remaining_inventory']=dat['AvailableQty'] - dat['prev_inventory_used_truck']
    return dat
```

```python
inv_df_list=[]
for sku in df['SKU'].unique():
    inv_df_list.append(get_inventory_remaining(df,sku))
df=pd.concat(inv_df_list)
df
```

```python
df=df[df['remaining_inventory']>=0]
df
```

```python
def getmaxRow(df,sku):
    valid_sku_type=sku+"_Min_QTY"
    df['sku_qty']=df['SKU']+"_"+df['Qty_Type']
    agg_df = df.groupby("row_id").agg({
    "sku_qty": list,  # Aggregate SKU into a list
    }).reset_index()

    agg_df['is_valid']=agg_df.apply(lambda row:1 if  valid_sku_type in row['sku_qty'] else 0, axis=1)
    print(agg_df)
    return agg_df.loc[agg_df['is_valid']==1,'row_id'].tolist()
```

```python
import numpy as np
done_data_list=[]
max_rows=10
row_filled=0
rows_to_ignore=[]
## Create Rows First
ran=max_skus
to_iterate=True
start_row_id=0
while to_iterate:
    for i in range(0,ran+1):
        filter_df=df[~df['index'].isin(rows_to_ignore)]
        sku_row_filled=0
        temp=filter_df[filter_df['Rank']==i]
        sku=temp['SKU'].tolist()[0]
        multiplier_order=temp['Multiplier'].tolist()[0]
        multiplier=multiplier_data[multiplier_order]
        same_multiplier_data=filter_df[filter_df['Multiplier']==multiplier_order]

        same_multiplier_data=same_multiplier_data.sort_values(by="index",ascending=True)
        cumulative_sum = [sum(multiplier[:i+1]) for i in range(len(multiplier))]

        same_multiplier_data['row_index']=same_multiplier_data.reset_index(drop=True).index+1
        same_multiplier_data['is_modulus']=same_multiplier_data['row_index'].apply(lambda x:1 if x in cumulative_sum else np.NaN)
        same_multiplier_data['cum_rows']=same_multiplier_data['is_modulus'].fillna(0).cumsum()
        same_multiplier_data['row_id']=same_multiplier_data.apply(lambda row:np.NaN if pd.isnull(row['is_modulus']) else row['cum_rows']+row_filled,axis=1 )
        same_multiplier_data['row_id']=same_multiplier_data['row_id'].bfill()
        
        
        print(same_multiplier_data)
        valid_row_id=getmaxRow(same_multiplier_data,sku)

        same_multiplier_data=same_multiplier_data[same_multiplier_data['row_id'].isin(valid_row_id)]
        same_multiplier_data=same_multiplier_data[same_multiplier_data['row_id']<=max_rows]
        done_data_list.append(same_multiplier_data)
        for val in same_multiplier_data['index'].tolist():
            rows_to_ignore.append(val)
        row_filled=max(same_multiplier_data['row_id'])
        if i==ran:
            to_iterate=False
        if row_filled==max_rows:
            to_iterate=False
        print('Row Filled',row_filled)
    



        


done_data=pd.concat(done_data_list)
done_data.to_csv("Done_Truck.csv",index=False)


    

    
```
